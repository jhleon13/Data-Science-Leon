```python
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
```

```python
def freq_table(dataset, index_freq):
    freq_table1 = {}
    for n in dataset:
        freq_counter = n[index_freq]
        if freq_counter in freq_table1:
            freq_table1[freq_counter] += 1
        else:
            freq_table1[freq_counter] = 1
    #for variable in freq_table1: # inpercentage form:
     #   freq_table1[variable] /= len(dataset)
      #  freq_table1[variable] *= 100
    return freq_table1
```

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

There are duplicate rows in android list. We will delete the rows with the dupicate, but keep the ones with the higher rating count. 

```python
#Dictionary with unique apps and the highest review count
reviews_max = {}

for n in android[1:]:
    name = n[0]
    n_reviews = float(n[3])
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max.update({name:n_reviews})
        # alternate script reviews_max[name] = n_reviews
        
    elif name not in reviews_max:
        reviews_max.update({name:n_reviews})

print("Expected length:", len(android[1:])-len(duplicate_apps))
print("Actual length:", len(reviews_max))
```

```python
def split(word):
    return [char for char in word]
def unknown_string(string):
    list_string = split(string)
    count1 = 0
    for n in list_string:
        if ord(n) > 127:
            count1 += 1
        if count1 > 3:
            return False
    return True
```

### For the ios apps I will look at the index value, index = -5, the prime_genre column. Accordinaly in the android dataset we will investigate the column with the index 1 for category, and index 9 for genre.

```python
# Making a Dictionary of a certain column results in Percentage
def freq_table(dataset, index_freq):
    freq_table1 = {}
    for n in dataset:
        freq_counter = n[index_freq]
        if freq_counter in freq_table1:
            freq_table1[freq_counter] += 1
        else:
            freq_table1[freq_counter] = 1
    for variable in freq_table1: # in percentage form:
        freq_table1[variable] /= len(dataset)
        freq_table1[variable] *= 100
    return freq_table1
```

```python
#The sorted() function doesn't work too well with dictionaries because it only considers and returns the dictionary keys.
# So we have to do this to sorted out
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
    
```

```python
display_table(free_ios_apps, -5)
```

```python
#Average number of user ratings 
genres_ios = freq_table(free_ios_apps, -5)#prime genre
table3 = {}
for genre in genres_ios:
    total = 0
    len_genre = 0
    for app in free_ios_apps:
        genre_app = app[-5]
        if genre_app == genre:            
            n_ratings = float(app[6])
            total += n_ratings
            len_genre += 1
    avg_n_ratings = total / len_genre
    table3[genre] = avg_n_ratings
#organize the last list
table_display = []
for key in table3:
    key_val_as_tuple = (table3[key], key)
    table_display.append(key_val_as_tuple)
        
table_sorted = sorted(table_display, reverse = True)
for entry in table_sorted:
    print(entry[1], ':', entry[0])
```

On average, navigation apps have the highest number of user reviews.

```python
#Average number of intalls per app genre
genres_android = freq_table(free_android_apps, 1)#Category
table4 = {}
for genre in genres_android:
    total = 0
    len_category = 0
    for app in free_android_apps:
        genre_app = app[1]
        if genre_app == genre:            
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            total += float(n_installs)
            len_category += 1
    avg_n_installs = total / len_category
    table4[genre] = avg_n_installs
    
#organize the last list
table_display = []
for key in table4:
    key_val_as_tuple = (table4[key], key)
    table_display.append(key_val_as_tuple)
        
table_sorted = sorted(table_display, reverse = True)
for entry in table_sorted:
    print(entry[1], ':', entry[0])
```
