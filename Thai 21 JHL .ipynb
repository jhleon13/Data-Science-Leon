```python
#Game start at 21 last flag(s) 1-3
import random
choices = [1,2,3]
start = 21
remainingflags = start
mod = len(choices) + 1
print('Starting amount of flags is =', '', start)

while remainingflags > 0:
    if remainingflags <= mod:
        while True:
            try:
                print('You can only pick:', choices)
                print('My pick is:')
                i = int(input())
                if (i == 1 or i == 2 or i == 3):
                    break
            except:
                continue 
        remainingflags = remainingflags - i            
        if remainingflags == 0:
            print("remainingflags=", remainingflags)
            print('You are so smart')
        else:
            print("remainingflags=", remainingflags)
            print("computerpicks =", remainingflags)
            remainingflags = remainingflags - remainingflags
            print("remainingflags=", remainingflags)
            print('You lost')
            break      
    else:
        while remainingflags > mod:
            while True:
                try:
                    print('You can only pick:', choices)
                    print('My pick is:')
                    i = int(input())
                    if (i == 1 or i == 2 or i == 3):
                        break
                except:
                    continue
            remainingflags = remainingflags - i
            mod4 = (remainingflags%mod)
            if mod4 == 0:
                print("remainingflags=",remainingflags)
                computerpicks = random.choice(choices)
                print("computerpicks =", computerpicks)
                remainingflags = remainingflags - computerpicks
                print("remainingflags=", remainingflags)
                continue
            else:
                print("remainingflags=",remainingflags)
                computerpicks = mod4
                print("computerpicks =", computerpicks)
                remainingflags = remainingflags - computerpicks
                print("remainingflags=", remainingflags)
                continue
```
